# 3. ARIMA Model

[ARIMA](https://www.machinelearningplus.com/time-series/arima-model-time-series-forecasting-python/), short for 'Auto Regressive Integrated Moving Average' is actually a class of models that 'explains' a given time series based on its own past values, that is, its own lags and the lagged forecast errors, so that equation can be used to forecast future values.

```python
X = year_wh['EVENT_ID'].values # We have 21 values 

model = ARIMA(X, order=(2,1,0))
model_fit = model.fit(disp=0)
print(model_fit.summary())
```

### Residuals

- PLotting the model residuals: suggesting that there may still be some trend information not captured by the model.

```python
residuals = pd.DataFrame(model_fit.resid)

residuals.plot(figsize=(16,6))
pyplot.show()
```

```python
print(residuals.describe()) 
```

Here we can see that there is a bias in the prediction (a non-zero mean in the residuals)

## ARIMA model for our Time Series

```python
size = int(len(X) * 0.66) # We train with 13 and test with 8
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()
for t in range(len(test)):
	model = ARIMA(history, order=(2,1,0)) 
	model_fit = model.fit(disp=0)
	output = model_fit.forecast(steps=2)[0]
	yhat = output[0]
	predictions.append(yhat)
	obs = test[t]
	history.append(obs)
	print('predicted=%f, expected=%f' % (yhat, obs))
```

```python
error_mse = mean_squared_error(test, predictions)
error_mae = mean_absolute_error(test, predictions)
root_mse = math.sqrt(error_mse)
```

# 4. Linear and Polynomial Regressions

- We know from plots that we don't have a linear rlation between time and number of events

```python
year_wh = year_wh.reset_index()

year_wh['YEAR']= year_wh['YEAR'].dt.year
year_wh['EVENTS']= year_wh['EVENT_ID']
```

```python
y = pd.DataFrame(year_wh['EVENTS'])
x = pd.DataFrame(year_wh['YEAR'])

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.33,random_state=42)

regr = linear_model.LinearRegression()

regr.fit(x_train,y_train)

y_pred = regr.predict(x_test)

regr.coef_

mse_lin = mean_squared_error(y_test, y_pred)

root_mse_lin = math.sqrt(mse_lin)


mae_lin = mean_absolute_error(y_test, y_pred)
```
